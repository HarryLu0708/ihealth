# Categorical Data Analysis

```python
# analysis for county_name

# we want to know how many people in these counties
population_distribution = data.groupby('county_name')['unique_member_cnt'].sum().reset_index()
population_distribution.rename(columns={'unique_member_cnt': 'total_population'}, inplace=True)

population_distribution
```

```python
plt.figure(figsize=(15, 10))  # 可以调整图的大小
sns.barplot(x='county_name', y='total_population', data=population_distribution)

plt.title('Population Distribution by County')
plt.xlabel('County Name')
plt.ylabel('Total Population')
plt.xticks(rotation=90)
plt.show()
```

```python
def cate_data_to_df_viz(target, idx_name, data):
    distribution = data.groupby(target)['unique_member_cnt'].sum().reset_index()
    distribution.rename(columns={"unique_member_cnt": 'total_population'}, inplace=True)
    plt.figure(figsize=(15, 10))  # 可以调整图的大小
    sns.barplot(x=idx_name, y='total_population', data=distribution)

    plt.title('Population Distribution by '+idx_name)
    plt.xlabel(idx_name)
    plt.ylabel('Total Population')
    plt.xticks(rotation=90)
    plt.show()
    return distribution
```

```python
# analysis for payer

# we want to know how many people in these counties
payer_distribution = data.groupby('payer')['unique_member_cnt'].sum().reset_index()
payer_distribution.rename(columns={'unique_member_cnt': 'total_population'}, inplace=True)

payer_distribution
```

```python
plt.figure(figsize=(10, 6))  # 可以调整图的大小
sns.barplot(x='payer', y='total_population', data=payer_distribution)

plt.title('Population Distribution by Payer')
plt.xlabel('Payer Name')
plt.ylabel('Total Population')
plt.xticks(rotation=45)
plt.show()
```

```python
# analysis for age_group
age_distribution = cate_data_to_df_viz("age_group_name","age_group_name",data)
```

```python
# viz for services
services_distribution = cate_data_to_df_viz("service","service",data)
```

```python
cols = num_data.columns
n_cols = 3
n_rows = len(cols) // n_cols if len(cols) % n_cols == 0 else (len(cols) // n_cols) + 1

plt.figure(figsize=(n_cols * 5, n_rows * 4))  # 调整总图的大小

for i, column in enumerate(cols, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.histplot(num_data[column], kde=True)
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```
