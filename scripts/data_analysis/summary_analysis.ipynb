```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\luyil\ihealth\datasets\preproc\decoded_data.csv")
data.head()
```

```python
data.dtypes
```

Vriables we will drop: 
- "Unnamed: 0": useless index number. 
- "row_id": another useless id data. 
- "svc_code" and "svc_code_type": we already decode these variables to "service" feature. 

```python
drop_data_1 = ["Unnamed: 0","row_id","svc_code","svc_code_type"]
dropped_data = data.drop(drop_data_1,axis=1)
dropped_data.head()
```

It very useful for us to know the proportion that insurer paid and member paid. 
- insurer_paid_prop = insurer_paid_amt / total_paid_amt
- member_paid_prop = member_paid_amt / total_paid_amt

```python
print(dropped_data.shape)

new_columns = [
    "insurer_paid_sum_prop",
    "insurer_paid_mean_prop",
    "insurer_paid_median_prop",
    "member_paid_sum_prop",
    "member_paid_mean_prop",
    "member_paid_median_prop"
]

upper_vars = [
    "insurer_paid_amt_sum",
    "insurer_paid_amt_mean",
    "insurer_paid_amt_median",
    "member_paid_amt_sum",
    "member_paid_amt_mean",
    "member_paid_amt_median"
]

lower_vards = [
    "total_paid_amt_sum",
    "total_paid_amt_mean",
    "total_paid_amt_median",
    "total_paid_amt_sum",
    "total_paid_amt_mean",
    "total_paid_amt_median"
]

for i in range(0, len(new_columns)):
    dropped_data[new_columns[i]] = dropped_data[upper_vars[i]] / dropped_data[lower_vards[i]]

print(dropped_data.shape)
```

```python
dropped_data.head()
```

We discovered some of the data are NA or NUll. 

```python
dropped_data.isna().mean()
```

```python
dropped_data.isnull().mean()
```

Approximate ~20% of the data are null for "insurer_paid_median_prop" and "member_paid_median_prop"; we choose not to use these features. For the rest of the data there're ~2% are null, we just directly drop those NA value. 

```python
# drop features
print(dropped_data.shape)
dropped_data_2 = dropped_data.drop(["insurer_paid_median_prop","member_paid_median_prop"],axis=1)
# drop NA
dropped_data_2 = dropped_data_2.dropna()
print(dropped_data_2.shape)
```

```python
dropped_data_2.isna().mean()
```

```python
dropped_data_2.head()
```

It's also will be helpful if we find the average operations per member by:
average_ops_cnt = clm_cnt / unique_member_cnt

```python
dropped_data_2["average_ops_cnt"] = dropped_data_2["clm_cnt"]/dropped_data_2["unique_member_cnt"]
dropped_data_2.isna().mean(), dropped_data_2.shape
```

Final drops:
- "svc_code_ctg": no description, we won't use it I guessed.  

```python
final_data = dropped_data_2.drop(["svc_code_ctg"],axis=1)
final_data.shape
```
