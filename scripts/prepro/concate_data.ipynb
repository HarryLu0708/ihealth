```python
import pandas as pd
import numpy as np
```

# Getting Data

```python
drugs = ["C:\\Users\\luyil\\ihealth\\datasets\\drugs\\rxdetail2009.xlsx"]

for i in range(10,21):
    f_name = f"C:\\Users\\luyil\\ihealth\\datasets\\drugs\\rxdetail20{i}.xlsx"
    print(f_name)
    drugs.append(f_name)

drugs
```

```python
services = ["C:\\Users\\luyil\\ihealth\\datasets\\services\\services2009.xlsx"]

for i in range(10,22):
    f_name = f"C:\\Users\\luyil\\ihealth\\datasets\\services\\services20{i}.xlsx"
    print(f_name)
    services.append(f_name)

services
```

```python
import re
```

```python
path = "C:\\Users\\luyil\\ihealth\\datasets\\services\\services2009.xlsx"
numbers = re.findall(r'\d+', path)
print(numbers[0])
```

```python
services_df = []
drugs_df = []

# read services files
for xlsx in services:
    year = re.findall(r'\d+', xlsx)
    sheet_name = year[0] + " Data"
    print(xlsx, sheet_name)
    df = pd.read_excel(xlsx,sheet_name=sheet_name)
    services_df.append(df)
```

```python
services_df[0].head()
```

```python
# read drugs files
for xlsx in drugs:
    year = re.findall(r'\d+', xlsx)
    sheet_name = year[0] + " Data"
    print(xlsx, sheet_name)
    df = pd.read_excel(xlsx,sheet_name=sheet_name)
    drugs_df.append(df)
```

```python
drugs_df[0].head()
```

# Concate Data

```python
# concate services data
combined_services_data = services_df[0]

for i in range(1,len(services_df)):
    print(i)
    combined_services_data = pd.concat([combined_services_data, services_df[i]],axis=0)

combined_services_data.head()
```

```python
# concate drugs data
combined_drugs_data = drugs_df[0]

for i in range(1,len(drugs_df)):
    print(i)
    combined_drugs_data = pd.concat([combined_drugs_data, drugs_df[i]],axis=0)

combined_drugs_data.head()
```

```python
combined_drugs_data.shape, combined_services_data.shape
```

# Save Files

```python
# included index
# save services file
combined_services_data.to_csv(r"C:\Users\luyil\ihealth\datasets\preproc\services_data.csv")
# save drugs file
combined_drugs_data.to_csv(r"C:\Users\luyil\ihealth\datasets\preproc\drugs.data.csv")
```
