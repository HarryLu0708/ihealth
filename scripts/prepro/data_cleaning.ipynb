# Services Data
## Missing Values

```python
services_data.describe()
```

```python
services_data.isnull().mean()
```

when we looked at the missing values and data, we found data can actually separated by two parts.

```python
service_df1 = services_data[~services_data["year"].isnull()]
service_df2 = services_data[services_data["year"].isnull()]
```

```python
service_df1.isnull().mean()
```

```python
service_df2.isnull().mean()
```

```python
service_df1 = service_df1.dropna(axis=1)
service_df1.isnull().mean()
```

```python
service_df1.head()
```

```python
service_df2 = service_df2.dropna(axis=1)
service_df2.isnull().mean()
```

```python
service_df2.head()
```

```python
service_df1.shape, service_df2.shape
```

```python
# save data files
service_df1.to_csv(r"C:\Users\luyil\ihealth\datasets\preproc\services\providers.csv")
service_df2.to_csv(r"C:\Users\luyil\ihealth\datasets\preproc\services\payers.csv")
```
