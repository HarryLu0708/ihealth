```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\luyil\ihealth\datasets\preproc\decoded_data.csv")
data.head()
```

```python
data.describe()
```

```python
data.dtypes
```

```python
data["avg_ops"] = data["clm_cnt"]/data["unique_member_cnt"]
data["avg_ops"]
```

```python
data["insurer_paid_prop"] = data["insurer_paid_amt_sum"]/data["total_paid_amt_sum"]
data["insurer_paid_prop"].head()
```

```python
data["member_paid_prop"] = data["member_paid_amt_sum"]/data["total_paid_amt_sum"]
data["member_paid_prop"].head()
```

```python
columns_we_can_drop = ["row_id","svc_code","svc_code_type"]
new_data = data.drop(columns_we_can_drop, axis=1)
new_data.head()
```

```python
new_data.isna().mean()
```

```python
new_data = new_data.dropna()
new_data.shape
```

```python
new_data.dtypes
```

```python
plt.figure(figsize=(10, 6))  
sns.boxplot(data=new_data)
plt.title('Boxplot of DataFrame')
plt.xticks(rotation=45) 
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
drop_cols = ["Unnamed: 0","county_code","age_group_code"]
```

```python
new_data = new_data.drop(drop_cols,axis=1)
new_data.shape
```

```python
new_data.dtypes
```

```python
scaler = MinMaxScaler()
cols = ["insurer_paid_amt_mean","insurer_paid_amt_sum","insurer_paid_amt_median",
        "total_paid_amt_mean","total_paid_amt_sum","total_paid_amt_median",
        "member_paid_amt_mean","member_paid_amt_sum","member_paid_amt_median",
        "unique_member_cnt", "clm_cnt", "avg_ops"
        ]
new_data[cols] = scaler.fit_transform(new_data[cols])
new_data[cols].head()
```

```python
plt.figure(figsize=(10, 6))  
sns.boxplot(data=new_data[cols])
plt.title('Boxplot of DataFrame')
plt.xticks(rotation=45) 
plt.show()
```
